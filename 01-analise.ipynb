```python
import pandas as pd
import geopandas as gpd
```

```python
df_lat = pd.read_csv('entrada/LATLONG.csv', sep=';')
```

```python
df_lat
```

```python
for col in ['LNG_LOJA', 'LAT_LOJA', 'LNG_FRIGO', 'LAT_FRIGO']:
    df_lat[col] = df_lat[col].str.replace(',', '.').astype(float)
df_lat
```

```python
input_dataframe = df_lat.copy()
lon_column_name = 'LNG_LOJA'
lat_column_name = 'LAT_LOJA'

geo_df_loja = gpd.GeoDataFrame(input_dataframe, geometry=gpd.points_from_xy(input_dataframe[lon_column_name], input_dataframe[lat_column_name]))
geo_df_loja
```

```python
input_dataframe = df_lat.copy()
lon_column_name = 'LNG_FRIGO'
lat_column_name = 'LAT_FRIGO'

geo_df_frigo = gpd.GeoDataFrame(input_dataframe, geometry=gpd.points_from_xy(input_dataframe[lon_column_name], input_dataframe[lat_column_name]))
geo_df_frigo
```

```python
# https://www.ibge.gov.br/geociencias/organizacao-do-territorio/malhas-territoriais/26565-malhas-de-setores-censitarios-divisoes-intramunicipais.html?=&t=downloads

geo_df = gpd.read_file('entrada/24SEE250GC_SIR.shp')
geo_df
```

```python
geo_df_frigo.set_crs(geo_df.crs, inplace=True)
geo_df_loja.set_crs(geo_df.crs, inplace=True)
```

```python
joined_geo_df_frigo = gpd.sjoin(geo_df_frigo, geo_df,  op='within')
joined_geo_df_loja = gpd.sjoin(geo_df_loja, geo_df, op='within')
joined_geo_df_frigo
```

```python
joined_geo_df_loja
```

```python
joined_geo_df_loja.columns
```

```python
df_loja = joined_geo_df_loja[['ID_left', 'LNG_LOJA', 'LAT_LOJA', 'index_right', 'ID_right', 'CD_GEOCODI', 'TIPO', 'CD_GEOCODB', 'NM_BAIRRO', 'CD_GEOCODS', 'NM_SUBDIST', 'CD_GEOCODD', 'NM_DISTRIT', 'CD_GEOCODM', 'NM_MUNICIP', 'NM_MICRO', 'NM_MESO']].copy()

df_loja
```

```python
df_renda = pd.read_excel('entrada/rendasetor.xlsx', dtype={'Cod_setor': str})
df_renda['V011'] = df_renda['V011'].str.replace(',', '.').astype(float)
df_renda
```

```python
df_loja = df_loja.merge(df_renda, left_on='CD_GEOCODI', right_on='Cod_setor', how='left')
```

```python
df_loja
```

```python
df_loja['CD_GEOCODI'].nunique()
df_loja.to_csv('saida/setores_censitarios_loja.csv', index=False)
```

# Distâncias

```python
from geopy.distance import geodesic

def vincenty_dist(lat1, lon1, lat2, lon2):
    point1 = (lat1, lon1)
    point2 = (lat2, lon2)
    return geodesic(point1, point2).kilometers

lat1, lon1 = 52.2296756, 21.0122287
lat2, lon2 = 41.8919300, 12.5113300

distance = vincenty_dist(lat1, lon1, lat2, lon2)
print(f"Exemplo distância (Vincenty): {distance} km")
```

```python
df_lat
```

```python
df_distancias = df_lat.copy()

df_distancias['DISTANCIA_KM'] = df_distancias.apply(lambda row: vincenty_dist(row['LAT_LOJA'], row['LNG_LOJA'], row['LAT_FRIGO'], row['LNG_FRIGO']), axis=1)
df_distancias
```

```python
df_distancias.to_csv('saida/distancias.csv', index=False)
```
